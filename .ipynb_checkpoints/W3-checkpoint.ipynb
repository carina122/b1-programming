```python
def greet(name):
    name[4] = 'b'
    
nome = ['C', 'a', 'r', 'i', 'n', 'a']
greet(nome)
print(nome)

# python functions are closed ==> when greet(nome) is called, 
# the var nome gets reassigned to a local (the function's) reference/pointer name which is inaccessible outside the function
# if you provide references to those new values w/o using globals, you will need to pass that back through return.
# when you pass an array or non primitive object into a function, 
# you can modify the object's attributes and have those modifications stick outside of the fct
# because that object itself contains the pointers to those values,
# therefore making them visible to anything else holding a pointer to that object.
```

```python
# a script that analyzes login attempts and identifies potential security threats

login_attempts = []

while True:
    username = input("Enter username (or 'quit' to stop): ")
    if username.lower() == 'quit':
        break
    status = input("Enter status: ")
    login_attempts.append((username, status))

print(login_attempts)

count = {}

for user, attempt in login_attempts:
    if attempt == 'failed':
        if user in count:
            count[user] += 1
        else: 
            count[user] = 1
            
for user in count:
    if count[user] >= 3:
        print ("Too many failed attempts! User " + user + " locked.")
    
```

```python
# port scanner that identifies open risky ports on network devices

devices = [("192.168.1.10", [22, 80, 443]),
           ("192.168.1.11", [21, 22, 80]), 
           ("192.168.1.12", [23, 80, 3389])]

risky_ports = [21, 23, 3389, 443]

counter = 0

print("Scanning network devices...")

for address, ports in devices:
    for port in ports:
        if port in risky_ports:
            print("WARNING: " + address + f' has risky port {port} open')
            counter += 1

if not counter:
    print("No security risks found")
else:
    print(f'{counter} security risks found')
```

```python
# password validator that checks compliance with security policies

passwords = ["Pass123",
             "SecurePassword1", "weak",
             "MyP@ssw0rd", "NOLOWER123"]

compliant, noncompliant = (0,0)

print("Validating passwords...")

def has_lowers(inputString):
    return any(char.islower() for char in inputString)

def has_uppers(inputString):
    return any(char.isupper() for char in inputString)

def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def failure_mssg(inputString, outputString):
    print("FAIL: " + inputString + outputString)
    return 1

for password in passwords:
    issues = 0
    if len(password) < 8:
        issues += failure_mssg(password, " - too short")
    if not has_lowers(password):
        issues += failure_mssg(password, " - no lowercase")
    if not has_uppers(password):
        issues += failure_mssg(password, " - no uppercase")
    if not has_numbers(password):
        issues += failure_mssg(password, " - no digit")
    if not issues:
        compliant += 1
        print("PASS: " + password + " - meets all requirements")
    else:
        noncompliant += 1

print('Summary: {} compliant, {} non-compliant'.format(compliant, noncompliant))
```
