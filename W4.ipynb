```python
# modular password validation tool
import string

def has_lowers(inputString):
    return any(char in string.ascii_lowercase for char in inputString)

def has_uppers(inputString):
    return any(char in string.ascii_uppercase for char in inputString)

def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def has_special_char(inputString):
    return any(char in string.punctuation for char in inputString)

def display_message(requirement, status):
    print(requirement + status)
    return int(status == '✓')

def det_status(statement):
    if statement:
        return '✓'
    else:
        return 'X'

def det_requirement(i):
    if (i == 0):
        return "8 characters" 
    elif (i == 1):
        return "one lowercase character"
    elif (i == 2):
        return "one uppercase character"
    elif (i == 3):
        return "one digit"
    else:
        return "special character"

def strength_check(length):
    if length == 8:
        return "low"
    elif length < 10:
        return "medium"
    elif length < 15:
        return "high"
    else:
        return "higher than The Empire State"

def validate_password(password):
    statements = [len(password) >= 8,
                  has_lowers(password),
                  has_uppers(password),
                  has_numbers(password),
                  has_special_char(password)]
    counter = 0
    print("Password needs to contain at least: ")
    for statement in statements:
        counter += display_message("-->" + det_requirement(statements.index(statement)) + ": ", det_status(statement))
    return counter == 5

while True:
    password = input("Type user password: ")
    if validate_password(password):
        print("Password strength: " + strength_check(len(password)))
        break
    print("Invalid password! Please try again.\n")
```
