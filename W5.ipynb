```python
# Parsing = reading the data
raw_data = """alice,success
bob,failed
alice,failed
diana,success
"""
# Split into lines
lines = raw_data.strip().split('\n')
print(lines)

login_attempts = {}
for line in lines:
    key, value = line.split(',')
    login_attempts[key] = login_attempts.get(key, 0) + 1

# get(key, val) function only returns val if the item doesn't exist - !!!!good for redundancy

print(login_attempts)
```

```python
# JSON - fancy files to efficiently store more complex data
import json

# Our data
user_scores = {
 "alice": 3,
 "bob": 5,
 "charlie": 2,
 "diana": 7
}

# Save to file
with open("scores.json", "w") as file:
 json.dump(user_scores, file, indent=4)
# The indent=4 parameter makes the JSON file readable by humans. Without it, everything appears on one line.
    
print("Data saved to scores.json")

# Load from file
with open("scores.json", "r") as file:
 loaded_scores = json.load(file)
    
print("Loaded data:", loaded_scores)
print("Type:", type(loaded_scores))

# Use the data normally
for user, score in loaded_scores.items():
 print(f"{user}: {score}")
    
# Modify and save again
loaded_scores["emma"] = 4
with open("scores.json", "w") as file:
 json.dump(loaded_scores, file, indent=4)
```

```python
# Student Grade Analyzer
import random
student_records = []
print("Enter student records")
for i in range(6):
    name = input("Name: ")
    score = random.randrange(1,101)
    print(f"Score: {score}")
    student_records.append((name, score))

scores = [score for name, score in student_records]  #<==>scores = [] / for name, score in student_records: scores.append(score) - line comprehension
highest = max(scores)
lowest = min(scores)
average = sum(scores) / len(scores)
print(f"\nCLASS STATISTICS\nHighest score: {highest}\nLowest score: {lowest}\nAverage score: {average:.2f}")

unique_scores = set(scores)
print(f"\nUNIQUE SCORES\n{unique_scores}\nTotal unique scores: {len(unique_scores)}")

grade_distribution = {}
for score in scores:
 grade_distribution[score] = grade_distribution.get(score, 0) + 1

print("\nGRADE DISTRIBUTION")
for score in sorted(grade_distribution.keys(), reverse=True):
 count = grade_distribution[score]
 plural = "students" if count != 1 else "student"
 print(f"Score {score}: {count} {plural}")
```

```python
# Personal expense tracker
import math
from collections.abc import Sequence

expense_records = []
category_totals = {}
amounts = []
print('Enter your expenses, bitte!')

for i in range(5):
    print(f'\nExpense {i + 1}')
    category = input("    --> category:")
    amount = float(input("    --> amount in €:"))
    date = input("    --> date(DD/MM/YYYY):")
    expense_records.append((category, amount, date))

for category,amount,date in expense_records:
    amounts.append(amount)
    category_totals[category] = category_totals.get(category,0) + amount

stats = {
    "Total Spending in €": sum(amounts), 
    "Average Expense": sum(amounts) / len(amounts),
    "Highest Expense": max(expense_records, key=lambda x: x[1]),  # find the record w the highest second element 
    "Lowest Expense": min(expense_records, key=lambda x: x[1])  # find the record w the highest second element
}

print("\nOVERALL SPENDING SUMMARY")
for att in stats.keys():
    if isinstance(stats[att], Sequence):   # check if a var is a complex data type; isinstance(var, tuple) can also be used but it's more restrictive
        category, amount, date = stats[att]
        print(f'{att}: {amount:.2f} (Category: {category}, Date: {date})')
    else:
        print(f'{att}: {stats[att]:.2f}')

unique_categories = set(category_totals.keys())
print(f'\nUNIQUE CATEGORIES SPENT ON\n{unique_categories}\nTotal unique categories: {len(unique_categories)}')

print("\nSPENDING BY CATEGORY")
for category,spending in category_totals.items():
    print(f'{category}: {spending}')
```
